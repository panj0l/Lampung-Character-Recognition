```python
def preprocess(img,input_size):
    nimg = img.resize(input_size, resample= 1)
    img_arr = (np.array(nimg))/255
    return img_arr
def reshape(imgs_arr):
    return np.stack(imgs_arr, axis=0)
```

```python
im = Image.open('model_learningspeed/pengenalan/data/aksara yi/yi.png')
X = preprocess(im,input_size)
X = reshape([X])
y = model.predict(X)
#Argmax adalah operasi yang menemukan argumen yang memberikan nilai maksimum dari fungsi target.
#Argmax paling sering digunakan dalam pembelajaran mesin untuk menemukan kelas dengan probabilitas prediksi terbesar.
print(labels[np.argmax(y)], np.max(y))
```
